Quanto o veículo rodou por ano?

```python
for index, row in dataset.iterrows():
    if (2019 - row['Ano'] != 0):
        #acessa pelo rotulo o index, e pelo index cria a nova coluna km_media, que vai calular a
        dataset.loc[index, 'Km_media'] = row['Quilometragem'] / (2019 - row['Ano'])
    else:
        dataset.loc[index,'Km_media'] = 0
        
dataset
```
